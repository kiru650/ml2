```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity

iris = load_iris()
X = iris.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

print("Cluster labels assigned to each sample:\n", labels)


sample1 = 20
sample2 = 60

print("\nSample 1 cluster:", labels[sample1])
print("Sample 2 cluster:", labels[sample2])

if labels[sample1] == labels[sample2]:
    print("Samples are similar (same cluster).")
else:
    print("Samples are NOT similar (different clusters).")

similarity_score = cosine_similarity(
    X_scaled[sample1].reshape(1, -1),
    X_scaled[sample2].reshape(1, -1)
)[0][0]

print("\nCosine similarity score:", similarity_score)
```
